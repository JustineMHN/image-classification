## Préparation des données avant l'entrainement
Pour réduire le temps de calcul et limiter les ressources utilisées pour l'entrainement, on procède par petits sous-ensemble (batch) dont la taille. De plus, il faut la plupart du temps redimensionner les images pour les mêmes raisons.

```python
batch_size = 16
img_height = 180
img_width = 180
```

### Ensembles d'apprentissage et de test
Ici les ensembles d'apprentissage et de test sont créés avec une fonction du module ***keras***.

```python
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

# Préparation des données d'apprentissage
train_ds = keras.preprocessing.image_dataset_from_directory(data_dir,
                                                            validation_split=0.2,
                                                            subset="training",
                                                            seed=123,
                                                            image_size=(img_height, img_width),
                                                            batch_size=batch_size)

# Préparation des données de validation
val_ds = keras.preprocessing.image_dataset_from_directory(data_dir,
                                                          validation_split=0.2,
                                                          subset="validation",
                                                          seed=123,
                                                          image_size=(img_height, img_width),
                                                          batch_size=batch_size)
```

### Normalisation de valeurs de pixels
Les valeurs des pixels sont normalisées en faisant une division par 255, 255 étant la valeur maximale d'un pixel codé sur 8 bits. L'objectif est d'avoir des valeurs entre 0 et 1.

```python
image_batch, labels_batch = next(iter(train_ds))
first_image = image_batch[0]
print("Valeurs des pixels avant normalisation", np.min(first_image), np.max(first_image))

# Création d'une couche de normalisation pour normaliser les valeurs de pixels entre 0 et 1
normalization_layer = layers.experimental.preprocessing.Rescaling(1./255)
normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]


# Les valeurs de pixels sont désormais compris entre 0 et 1 au lieu de 0 et 255
print("Valeurs des pixels après normalisation", np.min(first_image), np.max(first_image))
```

## Entraînement du modèle de classification
La classification est faite en utilisant un réseau de neurones avec des couches de convolution. Pour ce faire, on utilise package ***keras***. Le réseau est construit de façon séquentielle.

La fonction d'activation utilisée sur les couches cachées et la couche de sortie est `relu`. En plus des couches de convolutions, le réseau intègre également des couches de pooling.

```python
# Création d'un modèle séquentiel
num_classes = 5

model = Sequential([
    layers.experimental.preprocessing.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes)
])
```

La fonction `compile`permet de finaliser le réseau de neurones en indiquant la fonction d'optimisation, la fonction de perte et les métriques calculées à chaque itération.

```python
# Compilation du modèle
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Entrainement du réseau de neurones avec un ensemble de d'apprentissage, un ensemble de validation et le nombre d'epochs. Ce dernier correspond au nombre maximum d'itérations à faire pour mettre à jour les poids du réseau de neurones.

```python
# Entrainement du modèle
epochs=10
history = model.fit(train_ds, 
                    validation_data=val_ds,
                    epochs=epochs)
```

Une fois l'apprentissage on peut afficher les courbes correspondant aux différentes métriques "monitorées" durant la phase d'apprentissage. Elles permettent d'évaluer si l'entrainement est satisfaisant ou non.

```python
# Visualisation des métriques de la phase d'appentissage
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label="Métrique 'Accuracy' de l'apprentissage")
plt.plot(epochs_range, val_acc, label="Métrique 'Accuracy' de la validation")
plt.legend(loc='lower right')
plt.title("'Accuracy' des ensembles de validation et d'apprentissage")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="Fonction de perte de l'apprentissage")
plt.plot(epochs_range, val_loss, label="Fonction de perte de la validation")
plt.legend(loc='upper right')
plt.title("'Loss' des ensembles de validation et d'apprentissage")
plt.show()
```

### Augmentation de données
L'apprentissage précédent n'est pas satisfaisant. L'évolution de la fonction de perte sur l'ensemble de validation n'est pas celle attendue. Plusieurs raisons possibles à ce comportement :
- le manque de diversité dans la base d'apprentissage
- le nombre d'epochs n'est pas suffisant
- la normalisation des données n'est pas adaptée
- la fonction de perte n'est pas adaptée

Plusieurs solutions existent pour pallier ce problème. L'une d'elles est l'augmentation de données qui permet de créer de façon artificielle des variations dans la base existante afin d'y intégrer de la diversité. Dans le cas des images, on peut créer de nouvelles images en appliquant des transformations par exemple à celles existantes : rotation, translation, cropping, ... Il faut cependant s'assurer qu'on ne crée pas des données qui sont éloignées de la réalité afin de ne pas rajouter de nouveaux biais. 

```python
# Augmentation de données pour améliorer la qualité de l'apprentissage
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, img_width, 3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)

plt.figure(figsize=(10, 10))
for images, _ in train_ds.take(1):
  for i in range(9):
    augmented_images = data_augmentation(images)
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_images[5].numpy().astype("uint8"))
    plt.axis("off")
```

### Utilisation du dropout
La fonctionnalité "dropout" permet de désactiver certains neurones pendant un epoch d'apprentissage. Il suffit pour l'utiliser de rajouter une couche `Dropout` dans le réseau séquentiel en indiquant le ratio de neurones à désactiver.

```python
# Utilisation du dropout pour réduire le sur-apprentissage
new_model = Sequential([data_augmentation,
                    layers.experimental.preprocessing.Rescaling(1./255),
                    layers.Conv2D(16, 3, padding='same', activation='relu'),
                    layers.MaxPooling2D(),
                    layers.Conv2D(32, 3, padding='same', activation='relu'),
                    layers.MaxPooling2D(),
                    layers.Conv2D(64, 3, padding='same', activation='relu'),
                    layers.MaxPooling2D(),
                    # On rajoute ici une couche de Dropout
                    layers.Dropout(0.2),
                    layers.Flatten(),
                    layers.Dense(128, activation='relu'),
                    layers.Dense(num_classes)])

```

```python
# Compilation du modèle
new_model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# Nouvel entrainement du modèle
epochs=15
history = new_model.fit(train_ds, validation_data=val_ds, 
                        epochs=epochs)
```

```python
# Visualisation des métriques de la phase d'appentissage
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label="Métrique 'Accuracy' de l'apprentissage")
plt.plot(epochs_range, val_acc, label="Métrique 'Accuracy' de la validation")
plt.legend(loc='lower right')
plt.title("'Accuracy' des ensembles de validation et d'apprentissage")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="Fonction de perte de l'apprentissage")
plt.plot(epochs_range, val_loss, label="Fonction de perte de la validation")
plt.legend(loc='upper right')
plt.title("'Loss' des ensembles de validation et d'apprentissage")
plt.show()
```

Les courbes de la métrique `accuracy` et de la fonction de perte ont désormais une évolution conforme à ce qu'on attend. On est néanmoins bien loin d'avoir convergé. Il faut augmenter le nombre d'epochs tant que les deux courbes ne sont pas stables.
